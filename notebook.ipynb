## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Bagaimana tren pertumbuhan jumlah pemesanan setiap bulan pada tahun 2016-2018
- **Pertanyaan 2:**  kategori produk apa saja yang memiliki jumlah pemesanan yang banyak pada tahun 2016-2018
- **Pertanyaan 3:** daerah apa yang menjadi penyumbang pelanggan paling banyak pada tahun 2016-2018

### Cleaning Data

#### Fixing Data problem

````python
datatime_columns = ['order_purchase_timestamp',
                    'order_approved_at',
                    'order_delivered_carrier_date',
                    'order_delivered_customer_date',
                    'order_estimated_delivery_date']
for column in datatime_columns:
  df_Order[column] = pd.to_datetime(df_Order[column])
````

````python
df_Order.info()
````

````python
df_Order = df_Order[df_Order['order_status'] == 'delivered'].copy()
print('jumlah pesanan yang sudah diterima:', len(df_Order))
````

````python

df_Product = df_Product.merge(df_Product_Category, on='product_category_name', how='left')
df_Product['product_category_name_english'].fillna('others', inplace=True)
````

````python
df_Product.head()
````

````python
df_Geolocation.drop_duplicates(subset= 'geolocation_zip_code_prefix', inplace=True)
print('sisa duplikat pada geolocation:', df_Geolocation.duplicated().sum())
print('jumlah geolocation setelah didrop:', len(df_Geolocation))
````

````python
df_combined = df_Order.merge(df_Order_items, on='order_id', how='left')
df_combined = df_combined.merge(df_Product[['product_id', 'product_category_name_english']], on='product_id', how='left')
df_combined = df_combined.merge(df_Customer, on='customer_id', how='left')
df_combined = df_combined.merge(df_Geolocation[['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']],
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left')
df_combined['order_year_month'] = df_combined['order_purchase_timestamp'].dt.to_period('M').astype(str)
df_combined['revenue'] = df_combined['price'] + df_combined['freight_value']
````

## Exploratory Data Analysis (EDA)

### Explore Tren bulanan

````python
df_mountly = df_combined.groupby('order_year_month').agg(
    order_count=('order_id', 'nunique'),
    total_revenue=('revenue', 'sum')
).reset_index()
````

````python
df_mountly.head()
````

````python
df_mountly.describe()
````

````python
print('bulan paling top pemesanan :', df_mountly.loc[df_mountly['order_count'].idxmax(), 'order_year_month'])
print('bulan paling kurang pemesanan:', df_mountly.loc[df_mountly['order_count'].idxmin(), 'order_year_month'])
````

````python
df_Best_Region = df_combined.groupby('customer_state').customer_unique_id.nunique().sort_values(ascending=False).reset_index()
df_Best_Region.columns = ['customer_state', 'customer_count']
print('region yang terbaik:', df_Best_Region.loc[df_Best_Region['customer_count'].idxmax(), 'customer_state'])
df_Best_Region.head()
````

````python
df_category = df_combined.groupby('product_category_name_english').agg(total_orders=('order_id', 'count'),).reset_index().sort_values('total_orders', ascending=False)
print('kategori produk terbaik:', df_category.loc[df_category['total_orders'].idxmax(), 'product_category_name_english'])
df_category.head()
````

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana tren pertumbuhan jumlah pemesanan setiap bulan

````python
fig, axes = plt.subplots(1, 1, figsize=(12, 5))
fig.patch.set_facecolor('white')

x = range(len(df_mountly))

axes.set_facecolor('white')
axes.fill_between(x, df_mountly['order_count'], alpha=0.25, color='skyblue')
axes.plot(x, df_mountly['order_count'], color='Blue', linewidth=2, marker='o', markersize=4)

peak = df_mountly['order_count'].idxmax()
peak_pos = df_mountly.index.get_loc(peak)

axes.annotate(
    f"{df_mountly.loc[peak,'order_count']:,} pesanan",
    xy=(peak_pos, df_mountly.loc[peak,'order_count']),
    xytext=(max(0, peak_pos-2), df_mountly.loc[peak,'order_count'] * 0.9),
    arrowprops=dict(arrowstyle='->', color='black'),
    fontsize=10,color='black'
)

axes.set_xticks(list(x))
axes.set_xticklabels(df_mountly['order_year_month'], rotation=45, ha='right', fontsize=8)
axes.set_ylabel('Jumlah Pesanan', fontsize=11)
axes.set_title('Tren Pemesanan Per Bulan', fontsize=16, color='black', pad=10)
axes.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)
axes.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

### Pertanyaan 2: kategori produk apa saja yang memiliki jumlah pemesanan yang banyak


````python
top_orders = df_category.sort_values('total_orders', ascending=False).head(5)

fig, axes = plt.subplots(1, 1, figsize=(11, 7))
fig.patch.set_facecolor('#F8FBF9')

colors_l = ['#D00000' if i == 0 else '#FFB3B3' for i in range(len(top_orders))]

axes.set_facecolor('#F8FBF9')
bars1 = axes.barh(
    top_orders['product_category_name_english'][::-1],
    top_orders['total_orders'][::-1],
    color=colors_l[::-1],
    edgecolor='white'
)

for bar in bars1:
    w = bar.get_width()
    axes.text(
        w + 50,
        bar.get_y() + bar.get_height()/2,
        f'{int(w):,}',
        va='center',
        ha='left',
        fontsize=9,
        color='#7F0000'
    )

axes.set_title(
    'Top 5 Kategori berdasarkan Jumlah Pesanan\nPeriode 2016–2018',
    fontsize=12,
    color='#7F0000',
    pad=10,
    fontweight='bold'
)

axes.set_xlabel('Jumlah Pesanan', fontweight='bold')

axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)
axes.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

### Pertanyaan 3: daerah apa yang menjadi penyumbang pelanggan paling banyak

````python
fig, axes = plt.subplots(1, 1, figsize=(18, 7))
fig.patch.set_facecolor('#F8FBF9')

top_Region = df_Best_Region.head(5)
colors_s = ['#FFD60A' if i == 0 else '#FFF3B0' for i in range(len(top_Region))]
axes.set_facecolor('#F8FBF9')
bars = axes.barh(
    top_Region['customer_state'][::-1],
    top_Region['customer_count'][::-1],
    color=colors_s[::-1],
    edgecolor='white'
)

for bar in bars:
    w = bar.get_width()
    axes.text(
        w + 50,
        bar.get_y() + bar.get_height()/2,
        f'{int(w):,}',
        va='center',
        ha='left',
        fontsize=9,
        color='#7A5C00'
    )

axes.set_title(
    'Top 5 State berdasarkan Jumlah Pelanggan\nPeriode 2016–2018',
    fontsize=12,
    color='#7A5C00',
    pad=10,
    fontweight='bold'
)

axes.set_xlabel('Jumlah Pelanggan', fontweight='bold')

axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)
axes.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

## Conclusion & Recommendation

- **Conclusion pertanyaan 1:**
Berdasarkan analisis tren pemesanan bulanan pada periode 2016–2018, terlihat bahwa jumlah pesanan mengalami peningkatan yang cukup signifikan dari waktu ke waktu, meskipun terdapat fluktuasi di beberapa bulan. Pada awal periode, jumlah pesanan masih relatif rendah, namun mulai menunjukkan pertumbuhan yang konsisten sepanjang tahun 2017. Puncak pemesanan terjadi pada akhir tahun, khususnya sekitar November, yang mengindikasikan adanya pengaruh faktor musiman, seperti promosi besar atau event tertentu. Memasuki tahun 2018, jumlah pesanan cenderung stabil pada level tinggi meskipun sedikit mengalami penurunan di beberapa bulan. Hal ini menunjukkan bahwa bisnis telah mencapai fase pertumbuhan yang lebih matang, dan strategi penjualan sebaiknya difokuskan pada periode high season untuk memaksimalkan jumlah pesanan.
- **Conclusion pertanyaan 2:**
Berdasarkan analisis top 5 kategori produk dengan jumlah pesanan tertinggi pada periode 2016–2018, terlihat bahwa kategori bed_bath_table mendominasi dengan jumlah pesanan paling tinggi, diikuti oleh health_beauty, sports_leisure, furniture_decor, dan computers_accessories. Hal ini menunjukkan bahwa kebutuhan rumah tangga dan gaya hidup menjadi kontributor utama dalam penjualan.
- **Conclusion pertanyaan 3:**
Berdasarkan analisis top 5 state berdasarkan jumlah pelanggan pada periode 2016–2018, terlihat bahwa SP mendominasi secara signifikan dengan jumlah pelanggan tertinggi. Diikuti oleh RJ dan MG yang berada di posisi kedua dan ketiga, namun dengan selisih yang cukup besar dari SP. Sementara itu, RS dan PR melengkapi lima besar dengan jumlah pelanggan yang lebih rendah dibandingkan tiga state teratas.
